#Extracción

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

datos=pd.read_json('/content/TelecomX_Data.json')
datos
```

```python
datos_customer=pd.json_normalize(datos['customer'])
datos_customer
```

```python
datos_phone=pd.json_normalize(datos['phone'])
datos_phone
```

```python
datos_internet=pd.json_normalize(datos['internet'])
datos_internet
```

```python
datos_account=pd.json_normalize(datos['account'])
datos_account
```

```python
datos_customerID_Churn=datos[['customerID','Churn']]
```

```python
lista_datos=[datos_customerID_Churn,datos_customer, datos_phone,datos_internet, datos_account]
```

```python
df= pd.concat(lista_datos, axis=1)
df
```

#Transformación

```python
#Visualizar forma
df.shape
```

```python
#Visualizar datos, para tratar los que no coincidan
df.info()
```

###Manejo de inconsistencias

```python
#Buscar nulos
df.isnull().sum()

#No aparecen nulo, sin embargo al querer tratar columna Charges, aparecen espacios en blanco
```

```python
#Error de Charge.Total, determinar celda vacias
celdas_vacias = df[df['Charges.Total'].str.strip() == '']
print(celdas_vacias)
```

```python
#Convertir espacios en blanco a Nan ya que no se reconocen
df = df.replace([' ', ''], np.nan)

# Verificar si los espacios en blanco se han convertido a NaN
print(df.isnull().sum())
```

```python
#Llenar celdas con numero aletarios de 0 a 5000.0
import random
random.seed(123)
df['Charges.Total']=df['Charges.Total'].fillna(random.uniform(0, 5000.0))
```

```python
#Buscar nulos
df.isnull().sum()
```

```python
#Cambiar tipo de dato
df['Charges.Total']=df['Charges.Total'].astype(np.float64)
```

```python
#Llenar espacios en Churn
random.seed(123)

# Generar una respuesta aleatoria de "sí" o "no"
df['Churn']=df['Churn'].fillna(random.choice(["Yes", "No"]))
```

```python
df.isnull().sum()
```

```python
df['SeniorCitizen']=df['SeniorCitizen'].astype(np.bool)
```

```python
#Verificar tipos de datos
df.info()
```

```python
df[['Churn', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']] = df[['Churn', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']].replace({'Yes': True, 'No': False})
```

```python
#Verificar tipos de datos
df.info()
```

```python
df.columns
```

```python
#Renombrar columnas
df.rename(columns={
        'customerID': 'ID_cliente',
        'Churn': 'Cancelacion',
        'gender': 'Genero',
        'SeniorCitizen': 'Adulto_Mayor',
        'Partner': 'Asociado',
        'Dependents': 'Dependientes',
        'tenure': 'Antigüedad',
        'PhoneService': 'Servicio_Telefonico',
        'MultipleLines': 'Lineas_Multiples',
        'InternetService': 'Servicio_Internet',
        'OnlineSecurity': 'Seguridad_En_Linea',
        'OnlineBackup': 'Respaldo_En_Linea',
        'DeviceProtection': 'Proteccion_Dispositivo',
        'TechSupport': 'Soporte_Tecnico',
        'StreamingTV': 'TV_Streaming',
        'StreamingMovies': 'Peliculas_Streaming',
        'Contract': 'Contrato',
        'PaperlessBilling': 'Facturacion_Electronica',
        'PaymentMethod': 'Metodo_Pago',
        'Charges.Monthly': 'Cargos_Mensuales',
        'Charges.Total': 'Cargos_Totales'
    }, inplace=True)
```

```python
df
```

###Crear columna de cuentas diarias

```python
df['Cuentas_Diarias']=(df['Cargos_Mensuales']/30).round(3)
df
```

```python
#Extraer categorías a traducir del DF
columnas = ['Genero','Lineas_Multiples','Servicio_Internet', 'Seguridad_En_Linea', 'Respaldo_En_Linea','Proteccion_Dispositivo', 'Soporte_Tecnico', 'TV_Streaming','Peliculas_Streaming', 'Contrato', 'Metodo_Pago']
array = []

for columna in columnas:
    array.append(df[columna].unique())

# Crear un solo array
categorias = np.concatenate(array)

set(categorias)
```

```python
#Reemplazar textos
df.replace({
        'Bank transfer (automatic)': 'Transferencia bancaria (automatica)',
        'Credit card (automatic)': 'Tarjeta de credito (automatica)',
        'Electronic check': 'Cheque electronico',
        'Fiber optic': 'Fibra optica',
        'Mailed check': 'Cheque por correo',
        'Month-to-month': 'Mes a mes',
        'No internet service': 'Sin servicio de internet',
        'No phone service': 'Sin servicio telefonico',
        'One year': 'Un año',
        'Two year': 'Dos años',
        'Yes': 'Si','Female':'Femenino', 'Male':'Masculino'},inplace=True)
```

### DataFrame transformado y estandarizado

```python
#Guardar el DF con toda la transformación de datos
df.to_csv('DataFrame_Tratado.csv',index=False)
```
